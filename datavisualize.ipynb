```python
import re
import os
import matplotlib.pyplot as plt
import numpy as np

# 
output_dir = 'experiment_plots'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 
experiments = []
current_experiment = None

for line in d.strip().split('\n'):
    if line.startswith('####'):
        # 
        if current_experiment is not None:
            experiments.append(current_experiment)
        # 
        title = line.replace('####', '').strip()
        current_experiment = {
            'title': title,
            'epochs': [],
            'f1': [],
            'acc': []
        }
    elif current_experiment is not None and line.strip():
        #  epoch 
        # : epoch=0,loss=...,f1=...,acc=... epoch=1,...
        epoch_matches = re.findall(r'epoch=(\d+),loss=([\d.]+),f1=([\d.]+),acc=([\d.]+)', line)
        for epoch_match in epoch_matches:
            epoch, loss, f1, acc = epoch_match
            current_experiment['epochs'].append(int(epoch))
            current_experiment['f1'].append(float(f1))
            current_experiment['acc'].append(float(acc))

# 
if current_experiment is not None:
    experiments.append(current_experiment)

print(f" {len(experiments)} ")

# 
figures = []
for i, exp in enumerate(experiments):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    #  f1  acc 
    ax.plot(exp['epochs'], exp['f1'], color='blue', label='F1 Score', marker='o', linewidth=2, markersize=5)
    ax.plot(exp['epochs'], exp['acc'], color='red', label='Accuracy', marker='s', linewidth=2, markersize=5)
    
    # 
    ax.set_title(exp['title'], fontsize=12, fontweight='bold', pad=10)
    ax.set_xlabel('Epoch', fontsize=10)
    ax.set_ylabel('Score', fontsize=10)
    
    #  x 
    ax.set_xticks(exp['epochs'])
    
    # 
    ax.legend(loc='best', fontsize=9)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    #  y 
    y_min = min(min(exp['f1']), min(exp['acc'])) * 0.95
    y_max = max(max(exp['f1']), max(exp['acc'])) * 1.05
    ax.set_ylim(y_min, y_max)
    
    plt.tight_layout()
    
    # 
    safe_title = re.sub(r'[<>:"/\\|?*]', '_', exp['title'])  # 
    filename = f"{i+1:02d}_{safe_title}.png"
    filepath = os.path.join(output_dir, filename)
    plt.savefig(filepath, dpi=150, bbox_inches='tight')
    print(f": {filepath}")
    
    figures.append(fig)

# 
plt.show()

print(f"\n: {output_dir}/")
```
